```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "C:/Users/OI/Desktop/HeroesOfPymoli/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

#test read success

purchase_data.head(1)
```

```python
#get count for male purchases

male_count_df = purchase_data.Gender == "Male"

male_count = np.count_nonzero(male_count_df)

print("number of male purchases: " + str(male_count))

#get count for female purchases

female_count_df = purchase_data.Gender == "Female"

female_count = np.count_nonzero(female_count_df)

print("number of female purchases: " + str(female_count))

#get count of other/non-dislclosed purchases

other_count_df = purchase_data.Gender == "Other / Non-Disclosed"

other_count = np.count_nonzero(other_count_df)

print("number of other purchases: " + str(other_count))
```

```python
#get count for male PLAYERS

male_player_count_df = unique_player_interaction.Gender == "Male"

male_player_count = np.count_nonzero(male_player_count_df)

print("number of male PLAYERS: " + str(male_player_count))

#get count for female PLAYERS

female_player_count_df = unique_player_interaction.Gender == "Female"

female_player_count = np.count_nonzero(female_player_count_df)

print("number of female PLAYERS: " + str(female_player_count))

#get count of other/non-dislclosed PLAYERS

other_player_count_df = unique_player_interaction.Gender == "Other / Non-Disclosed"

other_player_count = np.count_nonzero(other_player_count_df)

print("number of other PLAYERS: " + str(other_player_count))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#find number of uniue items

number_of_items = purchase_data["Item ID"].nunique()

print(number_of_items)

#find average item price

average_item_price = purchase_data["Price"].mean()

print(round(average_item_price, 2))

#find total revenue

total_revenue = purchase_data["Price"].sum()

print(total_revenue)

#find total number of purchases

number_of_purchases = purchase_data["Price"].count()

print(number_of_purchases)
```

```python
#create dataframe to hold number of unique items, average price, and total revenue

purchase_analysis_dict = [{"Number of Unique Items": number_of_items, "Average Price": average_item_price, "Number of Purchases": number_of_purchases, "Total Revenue": total_revenue}]
purchase_analysis_df = pd.DataFrame(purchase_analysis_dict)


#format columns for dollars
purchase_analysis_df["Number of Unique Items"] = purchase_analysis_df["Number of Unique Items"]
purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map('${:,.2f}'.format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map('${:,.2f}'.format)

purchase_analysis_df = purchase_analysis_df.reset_index(drop=True)

#re-order columns
purchase_analysis_df = purchase_analysis_df[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

purchase_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins -keep in mind default is right bounded-
age_bins = [0,10,15,20,25,30,35,40,200]
#greate grouplables
age_groups = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

df = unique_player_interaction

#bin unique players and insert new column with bin labels

df["Age Groups"] = pd.cut(df["Age"], age_bins, labels=age_groups, right=False)

# group players by their new labels

player_ages_groups = df.groupby("Age Groups")

# creat a new df to hold results by count of age groups

player_ages_counts = player_ages_groups["Age Groups"].count()

# player_ages_counts

player_age_percentages = player_ages_counts.divide(number_of_players)*100

# player_age_percentages
```

```python
#build data frame for age demographics
age_demographics_df = pd.DataFrame({"Total Count": player_ages_counts, "Percentage of Players": player_age_percentages})

#round percent to two decimals
decimals = 2
age_demographics_df["Percentage of Players"]= age_demographics_df["Percentage of Players"].apply(lambda x: round(x,decimals))

#display dataframe

age_demographics_df
```

```python
#bin purchase data
purchase_data["Age Groups"] = pd.cut(purchase_data["Age"], age_bins, labels=age_groups, right=False)

#groupby

purchase_data_agegroup_groups = purchase_data.groupby(["Age Groups"])

###get counts by age group

purchase_data_age_counts = purchase_data_agegroup_groups["SN"].count()
# purchase_data_age_counts

###get average purchase by age group

purchase_data_age_avgprice = purchase_data_agegroup_groups["Price"].mean()
# purchase_data_age_avgprice

###get total purchase value by agegroup

purchase_data_age_total = purchase_data_agegroup_groups["Price"].sum()
# purchase_data_age_total
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# count number of purchases made by each individual

sales_by_individual = purchase_data["SN"].value_counts()
#test
# sales_by_individual

#---------------------------------------
# group purchase data by screen name AND purchase id
SN_PurchID_grp = purchase_data.groupby(["SN","Purchase ID"])

# make dataframe to hold sum of price column for individual and purchase id groupby
SN_PurchID_grp_df = SN_PurchID_grp["Price"].sum()

# run SECOND group by for individaul screenname
SN_group = SN_PurchID_grp_df.groupby(["SN"])

#----------------------------------------
#make dataframe to hold average spent by individual
SN_group_df = SN_group.mean()

#make dataframe to hold total spent by individual
SN_group_df_sum = SN_group.sum()

#test
# SN_group_df.head()

# Make df to report top spenders
top_spenders_df = pd.DataFrame({"Purchase Count":sales_by_individual, "Average Purchase Price": SN_group_df, "Total Purchase Value": SN_group_df_sum}) 

#sort values in output by total spent
top_spenders_df = top_spenders_df.sort_values("Total Purchase Value", ascending=False)

#format columns for dollars
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map('${:,.2f}'.format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map('${:,.2f}'.format)

#test
top_spenders_df.head()
```
